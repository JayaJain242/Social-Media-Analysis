```python
##Insights
1. Instagram is the most preferred platform for posting content, which means higher user activity compared to other platforms.
2. User activity is highest on weekend,especially Saturday which means people are more active on social media during their free time.
3. The average number of likes per post is quite high whioch indicates strong user engagement across the dataset.
4. There is a very weak correlation between followers and likes,so more followers does not always mean higher engagement.
5. The distribution of posts across platforms shows a clear dominance of Instagram.
```
